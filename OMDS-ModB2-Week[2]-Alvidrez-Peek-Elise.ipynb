```python
# DATASET #1: Ecommerce Customer Churn Analysis and Prediction

# STEP 1: Read File
# Load the CSV file for this dataset.
# This dataset has customer-level information related to e-commerce behavior,
# satisfaction, complaints, purchase activity, and whether the customer churned.
churn_df = pd.read_excel("D1-ecommerce-customer-churn.xlsx")


# STEP 2: First Few Rows
# This uses head() to look at the first few rows of the churn dataset. (Dataset #1)
# Note: A value of 1 means the customer churned, and a value of 0 means the customer did not churn.
print("Ecommerce Customer Churn dataset:")
display(churn_df.head())


```

```python
# STEP 3: Dataset Shape (Dataset #1)
# The shape shows the number of rows and columns.

print("Ecommerce Customer Churn dataset shape:")
print(churn_df.shape)
print()

# STEP 4: info() is used to see the column names, data types, and non-missing counts. (Dataset #1)

print("Ecommerce Customer Churn dataset info:")
print()
churn_df.info()

# Important observations:
# The dataset has 5630 rows and 20 columns.
# Some columns have missing values because their non-null count is less than 5630.
print()
print("Observation:")
print("The dataset has 5630 rows and 20 columns.")
print("Some columns have missing values, so the data needs some cleaning before using it.")
print("The average churn value is about 0.17, which means around 17% of customers churned.")


```

```python
# STEP 5: I used describe() to get summary statistics for numeric columns. (Dataset #1)

print("Ecommerce Customer Churn dataset summary:")
display(churn_df.describe())

# Important observation:
# Some columns have averages that are easy to understand.
# For example, the average churn is about 0.17. So most customers did not churn.
# The average satisfaction score is around 3, which is in the middle of the 1 to 5 scale.
# The median tenure is 9, meaning half of customers have tenured below 9 and half are above 9.

```

```python
# STEP 6: Check Duplicates (Dataset #1)
# Duplicate rows could mean the same customer record was accidentally repeated.

churn_duplicate_rows = churn_df.duplicated().sum()

print("Number of duplicate rows in the Ecommerce Customer Churn dataset:")
print(churn_duplicate_rows)

# I created a clean copy with duplicate rows removed.
# This does not change the original data.

churn_clean_df = churn_df.drop_duplicates()

print("Churn dataset shape after removing duplicate rows:")
print(churn_clean_df.shape)

# I checked for duplicate values in CustomerID.
# CustomerID should be unique because each customer should only have one ID.

duplicate_customer_ids = churn_df["CustomerID"].duplicated().sum()

print("Number of duplicate CustomerID values:")
print(duplicate_customer_ids)

print()

# If there are duplicated CustomerID values, this will show some examples.

if duplicate_customer_ids > 0:
    print("Examples of duplicated CustomerID rows:")
    display(churn_df[churn_df["CustomerID"].duplicated(keep=False)].head())

# Important observation:
# I checked for duplicate rows and duplicate CustomerID values.
# If CustomerID had duplicates, that could mean the same customer was listed more than once.

```

```python
# STEP 7: Mean, median, and mode for numeric columns. (Dataset #1)
#
# I calculated the mode because describe() does not include it while it does include mean median and mode.
# The mode is useful because it shows the value that appears most often.
# For example, if the Churn column has a mode of 0, that means most customers
# did not churn. This can also help reveal class imbalance.

# I selected only numeric columns because mean and median only work properly
# with numeric data.
numeric_columns = churn_df.select_dtypes(include=np.number).columns

# This calculates the average value for each numeric column.
mean_values = churn_df[numeric_columns].mean()

# This calculates the middle value for each numeric column.
median_values = churn_df[numeric_columns].median()

# This calculates the most common value for each numeric column.
# mode() can return more than one value if there is a tie.
# iloc[0] takes the first mode value so it fits neatly into the summary table.
mode_values = churn_df[numeric_columns].mode().iloc[0]

# This combines the mean, median, and mode into one table.
summary_stats = pd.DataFrame({
    "mean": mean_values,
    "median": median_values,
    "mode": mode_values
})

# This displays the table rounded to 2 decimal places.
print("Mean, median, and mode for numeric columns:")
display(summary_stats.round(2))

print("Observation:")
print("The average Churn value is about 0.17, so most customers did not churn.")
print("The median Churn value is 0, which also shows that most customers stayed.")
print("The average SatisfactionScore is around 3, so satisfaction looks pretty average.")
```

```python
# STEP 8: Missing values in each column.
# I checked for missing or null values in each column.
# Missing values are blank values in the dataset.

missing_count = churn_df.isnull().sum()
missing_percent = churn_df.isnull().mean() * 100

missing_table = pd.DataFrame({
    "missing_count": missing_count,
    "missing_percent": missing_percent
})

missing_table = missing_table.sort_values("missing_count", ascending=False)

print("Missing values in each column:")
print()
display(missing_table.round(2))

print()
print("Observation:")
print("Some columns have missing values.")
print("The columns with the most missing values include DaySinceLastOrder, OrderAmountHikeFromlastYear, Tenure, OrderCount, CouponUsed, HourSpendOnApp, and WarehouseToHome.")
print("The dataset is still usable because most of the data is there.")
```

```python
# STEP 9: Fill or Remove Missing Data (Dataset #1)
# This keeps the original data from accidentally getting changed.

# I made a copy of the dataset before changing anything.

churn_clean_df = churn_df.copy()

# I filled missing numeric values with the median.
# I chose the median because it is simple and not affected as much by very large values.

numeric_columns_with_missing = churn_clean_df.select_dtypes(include=np.number).columns[
    churn_clean_df.select_dtypes(include=np.number).isnull().sum() > 0
]

for column in numeric_columns_with_missing:
    churn_clean_df[column] = churn_clean_df[column].fillna(churn_clean_df[column].median())

# I filled missing text/category values with "Unknown".
# This is useful if there are missing values in word columns.

category_columns_with_missing = churn_clean_df.select_dtypes(exclude=np.number).columns[
    churn_clean_df.select_dtypes(exclude=np.number).isnull().sum() > 0
]

for column in category_columns_with_missing:
    churn_clean_df[column] = churn_clean_df[column].fillna("Unknown")

print("Missing values after cleaning:")
print(churn_clean_df.isnull().sum().sum())

print()
print("Observation:")
print("I filled missing numeric values with the median.")
print("I filled missing category values with Unknown if there were any.")
print("I did not remove the rows because there were not too many missing values, and I wanted to keep as much data as possible.")
```

```python
# STEP 10: Check for inconsistent data or bad data that does not make sense (Dataset #1)
# I checked for values that do not make sense.
# Churn should only be 0 or 1.
# Complain should only be 0 or 1.
# SatisfactionScore should be between 1 and 5.
# CityTier should be between 1 and 3.
# Some columns should not have negative values.

bad_churn = churn_clean_df[~churn_clean_df["Churn"].isin([0, 1])]

bad_complain = churn_clean_df[~churn_clean_df["Complain"].isin([0, 1])]

bad_satisfaction = churn_clean_df[
    (churn_clean_df["SatisfactionScore"] < 1) |
    (churn_clean_df["SatisfactionScore"] > 5)
]

bad_city_tier = churn_clean_df[
    (churn_clean_df["CityTier"] < 1) |
    (churn_clean_df["CityTier"] > 3)
]

negative_tenure = churn_clean_df[churn_clean_df["Tenure"] < 0]

negative_cashback = churn_clean_df[churn_clean_df["CashbackAmount"] < 0]

negative_orders = churn_clean_df[churn_clean_df["OrderCount"] < 0]

print("Rows with bad Churn values:")
print(bad_churn.shape[0])

print("Rows with bad Complain values:")
print(bad_complain.shape[0])

print("Rows with bad SatisfactionScore values:")
print(bad_satisfaction.shape[0])

print("Rows with bad CityTier values:")
print(bad_city_tier.shape[0])

print("Rows with negative Tenure:")
print(negative_tenure.shape[0])

print("Rows with negative CashbackAmount:")
print(negative_cashback.shape[0])

print("Rows with negative OrderCount:")
print(negative_orders.shape[0])

print()
print("Observation:")
print("These checks help find values that would not make sense.")
print("For example, Churn should only be 0 or 1, and SatisfactionScore should stay between 1 and 5.")
print("None of these counts are above 0, so doesnt look problematic.")
```

```python
# Step 11: Check for outliers
print("Largest WarehouseToHome values:")
print()
display(churn_clean_df[["CustomerID", "WarehouseToHome"]].sort_values("WarehouseToHome", ascending=False).head())

print()
print("Largest DaySinceLastOrder values:")
print()
display(churn_clean_df[["CustomerID", "DaySinceLastOrder"]].sort_values("DaySinceLastOrder", ascending=False).head())

print()
print("Largest CashbackAmount values:")
print()
display(churn_clean_df[["CustomerID", "CashbackAmount"]].sort_values("CashbackAmount", ascending=False).head())

print()
print("Observation:")
print("Some values are much higher than most of the others.")
print("These might be real, but I would check them more carefully before doing a serious analysis.")
```

```python
#Step 12: Check for Class imbalance (Dataset #1)
# Class imbalance means one category has way more rows than another.
# I check the Churn column because that is the main target column.

churn_counts = churn_clean_df["Churn"].value_counts().sort_index()
churn_percent = churn_clean_df["Churn"].value_counts(normalize=True).sort_index() * 100

churn_balance_table = pd.DataFrame({
    "count": churn_counts,
    "percent": churn_percent
})

print("Churn class balance:")
print()
display(churn_balance_table.round(2))

print()
print("Observation:")
print("There is class imbalance in the Churn column.")
print("Most customers did not churn.")
print("This matters because a model might learn the non-churn group better than the churn group.")
```

```python
# Step 13: Basic Chart for Churn (Dataset #1)

# This graph shows how many customers churned and how many did not churn.
# Churn = 0 means the customer did not leave.
# Churn = 1 means the customer left.

churn_counts_for_plot = churn_clean_df["Churn"].value_counts().sort_index()

plt.figure(figsize=(7, 5))

plt.bar(
    ["Did Not Churn", "Churned"],
    churn_counts_for_plot.values,
    color=["skyblue", "salmon"]
)

plt.title("Dataset #1: Customer Churn Counts")
plt.xlabel("Customer Status")
plt.ylabel("Number of Customers")
plt.grid(axis="y", alpha=0.3)

plt.show()

print("Observation:")
print("This graph shows that many more customers did not churn than churned.")
print("The churned group is smaller, but it is still important because those are the customers the company wants to understand.")
```

```python
# Step 14: Compares average Tenuure (length with company) for customers who churned/not churn (Dataset #1)
# This graph compares average tenure for customers who churned and did not churn.
# Tenure means how long the customer has been with the company.

average_tenure_by_churn = churn_clean_df.groupby("Churn")["Tenure"].mean()

plt.figure(figsize=(7, 5))

plt.bar(
    ["Did Not Churn", "Churned"],
    average_tenure_by_churn.values,
    color=["skyblue", "salmon"]
)

plt.title("Dataset #1: Average Tenure by Churn Status")
plt.xlabel("Customer Status")
plt.ylabel("Average Tenure")
plt.grid(axis="y", alpha=0.3)

plt.show()

print("Observation:")
print("This graph helps compare customers who stayed and customers who left.")
print("If churned customers have lower average tenure, then newer customers may be more likely to leave.")
```

```python
# STEP 8: Fix number columns that are stored as text

# I made a copy of the dataset first to clean the copy without changing the original.

behavior_clean_df = behavior_df.copy()

# Then I cleaned the Sales column.
# I changed the values to strings just in case there are mixed data types.
# Then I removed the dollar sign.
# Then I removed commas.
# Finally, pd.to_numeric changes the cleaned text into numbers.

behavior_clean_df["Sales_num"] = pd.to_numeric(
    behavior_clean_df["Sales"].astype(str).str.replace("$", "", regex=False).str.replace(",", "", regex=False),
    errors="coerce"
)

# Profit also has dollar signs, so I remove the dollar signs and commas.
# Then I converted it into a number and save it as a new column called Profit_num.

behavior_clean_df["Profit_num"] = pd.to_numeric(
    behavior_clean_df["Profit"].astype(str).str.replace("$", "", regex=False).str.replace(",", "", regex=False),
    errors="coerce"
)

# This column also has dollar signs, so it needs the same cleaning.
# I saved the cleaned version as Shipping_Cost_num.

behavior_clean_df["Shipping_Cost_num"] = pd.to_numeric(
    behavior_clean_df["Shipping Cost"].astype(str).str.replace("$", "", regex=False).str.replace(",", "", regex=False),
    errors="coerce"
)

# Quantity should also be a number.


behavior_clean_df["Quantity_num"] = pd.to_numeric(
    behavior_clean_df["Quantity"],
    errors="coerce"
)

# Discount should also be numeric.
# I convert it into a number so I could use it in calculations.

behavior_clean_df["Discount_num"] = pd.to_numeric(
    behavior_clean_df["Discount"],
    errors="coerce"
)

print("Checking the cleaned numeric columns:")
print()

display(behavior_clean_df[[
    "Sales",
    "Sales_num",
    "Profit",
    "Profit_num",
    "Shipping Cost",
    "Shipping_Cost_num",
    "Quantity",
    "Quantity_num",
    "Discount",
    "Discount_num"
]].head())

print()
```

```python
# STEP 11: Fill or Remove Missing Data

# I make another copy before filling missing values.
# This keeps the original version safe.

behavior_filled_df = behavior_clean_df.copy()

# I fill missing Region values with Unknown because Region is a category.

behavior_filled_df["Region"] = behavior_filled_df["Region"].fillna("Unknown")

# I fill missing Order Priority values with Unknown because it is also a category.

behavior_filled_df["Order Priority"] = behavior_filled_df["Order Priority"].fillna("Unknown")

# I fill missing Quantity_num values with the median.
# I use the median because it is simple and works well if there are unusual values.

behavior_filled_df["Quantity_num"] = behavior_filled_df["Quantity_num"].fillna(
    behavior_filled_df["Quantity_num"].median()
)

print("Missing values after cleaning:")
print(behavior_filled_df.isnull().sum().sum())

print()
print("Observation:")
print("I filled missing category values with Unknown to be safe.")
print("I filled the missing quantity value with the median.")
print("I did not remove rows because there were not many missing values.")
```

```python
# STEP 12: Check for inconsistent data or bad data that does not make sense

# I check for values that do not make sense.
# Age should not be negative.
# Quantity should be greater than 0.
# Discount should not be negative.
# Sales and shipping cost should not be negative.
# Like, Share, and Add to Cart should only be 0 or 1.

bad_age = behavior_filled_df[
    (behavior_filled_df["Age"] < 0) |
    (behavior_filled_df["Age"] > 120)
]

bad_quantity = behavior_filled_df[
    behavior_filled_df["Quantity_num"] <= 0
]

bad_discount = behavior_filled_df[
    (behavior_filled_df["Discount_num"] < 0) |
    (behavior_filled_df["Discount_num"] > 1)
]

bad_sales = behavior_filled_df[
    behavior_filled_df["Sales_num"] < 0
]

bad_shipping = behavior_filled_df[
    behavior_filled_df["Shipping_Cost_num"] < 0
]

bad_like = behavior_filled_df[~behavior_filled_df["Like"].isin([0, 1])]
bad_share = behavior_filled_df[~behavior_filled_df["Share"].isin([0, 1])]
bad_cart = behavior_filled_df[~behavior_filled_df["Add to Cart"].isin([0, 1])]

print("Rows with bad Age values:")
print(bad_age.shape[0])

print("Rows with bad Quantity values:")
print(bad_quantity.shape[0])

print("Rows with bad Discount values:")
print(bad_discount.shape[0])

print("Rows with negative Sales values:")
print(bad_sales.shape[0])

print("Rows with negative Shipping Cost values:")
print(bad_shipping.shape[0])

print("Rows with bad Like values:")
print(bad_like.shape[0])

print("Rows with bad Share values:")
print(bad_share.shape[0])

print("Rows with bad Add to Cart values:")
print(bad_cart.shape[0])

print()
print("Observation:")
print("These checks help find values that would not make sense.")
print("For example, Like, Share, and Add to Cart should only be 0 or 1.")
print("These are turned out to be 0")
```

```python
# Dataset #1

# This is my rough Storytelling With Data graph.
# I am using a simple bar chart and highlighting the churned customers.
# This is similar to the book because it uses color to show the important point.

churn_story_df = churn_clean_df["Churn"].value_counts().sort_index().reset_index()
churn_story_df.columns = ["Churn", "count"]

colors = []

for churn_value in churn_story_df["Churn"]:
    if churn_value == 1:
        colors.append("red")
    else:
        colors.append("gray")

fig, ax = plt.subplots(figsize=(8, 5))

ax.bar(
    ["Did Not Churn", "Churned"],
    churn_story_df["count"],
    color=colors
)

ax.set_title("Dataset #1: Most customers in this dataset did not churn")
ax.set_xlabel("Customer Status")
ax.set_ylabel("Number of Customers")

# I remove the top and right borders to make the graph cleaner.

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# I add a note to point out the churned customer group.

churned_count = churn_story_df[churn_story_df["Churn"] == 1]["count"].iloc[0]

ax.annotate(
    "Churned customers",
    xy=(1, churned_count),
    xytext=(0.50, churned_count * 1.6),
    arrowprops={"arrowstyle": "->", "color": "red"},
    color="red"
)

plt.grid(axis="y", alpha=0.3)
plt.show()

print("Observation:")
print("This graph makes the main point easy to see.")
print("Most customers did not churn, but the churned group is highlighted because it is the group the company probably cares about most.")
```
